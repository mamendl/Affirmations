Se intenta predecir la actividad que estaba realizando un paciente a partir de distinta información como el nivel de azúcar o el pulso.

0- De pie 1- Caminando 2- Sentado 3- Cayendo 4- Con calambres 5- Corriendo

Las columnas indican:

SL: nivel de azucar

EEG: ratio monitor EEG

BP: presión en sangre

HR: ratio de latidos

CIRCULATION: circulación de sangre

```python
traindf = pd.read_csv("train.csv")
sampledf = pd.read_csv("sample.csv")
testdf = pd.read_csv("test.csv")
```

```python
traindata = traindf.values
```

```python
descriptores,clases = traindata[:,0:-1],traindata[:,-1]
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(descriptores,clases)
```

```python
testdata = testdf.values
```

```python
#predicciones = knn.predict(testdata)
```

```python
modeloKNN = KNeighborsClassifier(n_neighbors=6)
modeloKNN.fit(descriptores,clases)
```

```python
def evalua(clasificador,testData=testdata,testLabels=clases):
    prediccion = clasificador.predict(testData)
    correctas = 0
    for i,z in zip(prediccion, testLabels):
      if (i == z):
        correctas += 1
    return correctas
```

```python
evalua(modeloKNN)
```

```python
evalua(knn)
```

```python
predicciones = modeloKNN.predict(testdata)
```

```python
soldf = pd.DataFrame([(i+1,int(pred)) for i, pred in enumerate(predicciones)])
```

```python
soldf.columns = ['Id','Prediction']
```
